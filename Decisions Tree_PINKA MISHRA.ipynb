Importing library

# Builing Model

```python
from sklearn.model_selection import train_test_split
x = df.iloc[:,:3]
y = df['species']
x_train , x_test , y_train , y_test = train_test_split(x ,y ,test_size = 0.30, random_state=100)
```

fit decision tree to train data and predict model

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dt= DecisionTreeClassifier()
dt.fit(x_train, y_train)

pt=dt.predict(x_test)
```

```python
## print metric to get the accurecy of model
print("Accuracy: ",dt.score(x_test,y_test)*100)
```

# Visualize Decision Tree

```python
from sklearn import tree

fig = plt.figure(figsize=(25,20))
_ = tree.plot_tree(dt,
                   feature_names=['sepal length','sepal width','petal length','petal width' ],
                   class_names=['Setosa','Versicolor','Virginica'],
                   filled=True )
```

```python
from sklearn.metrics import accuracy_score
print(" Decision Tree Accuracy: ",accuracy_score(y_test,pt))
```

Decision Tree has gave an Accuracy of 93.33%.
